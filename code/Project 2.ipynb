```python
def make_system(ninja, share_rate):
    init = State(N=90, E=0, F=0, A=0, D=0)
    
    t0 = 0 
    t_end = 30
    
    return System(init=init, t0=t0, t_end=t_end, ninja=ninja, share_rate=share_rate)
```

```python
def update_func(state, system):
    n, e, f, a, d = state
    
    share_rate1 = 0.5 
    share_rate2 = 0.6
    share_rate3 = 0.7 
    share_rate4 = 0.8 
    
    
    exposed = (n*ninja) + (n * share_rate1)
    familiar = (e * ninja) + (e * share_rate2)
    admirer = (f * ninja) + (f * share_rate3)
    devote = (a*ninja) + (a * share_rate4)
    
    n -= exposed
    e += exposed - familiar
    f += familiar - admirer 
    a += admirer - devote 
    d += devote 
    
    return State(N=n, E=e, F=f, A=a, D=d)
```

```python
def run_simulation(system, update_func):
    
    catchy = system.init 
    for t in linrange(system.t0, system.t_end):
        catchy = update_func(init, system)
    return catchy
```

```python
song = make_system(ninja, share_rate)
run_simulation(song, update_func)
```

```python
def run_simulation(system, update_func):
    unpack(system)
    
    frame = TimeFrame(columns=init.index)
    frame.loc[system.t0] = init
    
    for t in linrange(t0, t_end):
        frame.loc[t+1] = update_func(system, frame.loc[t])
        
    system.results = frame
```

```python
run_simulation(song, update_func)
```

```python
def plot_results(N, E, F, A, D):
    plot(N, '--', color='blue', label='Never Heard')
    plot(E, '-', color='red', label='Exposed')
    plot(F, ':', color='green', label='Familiar')
    plot(A, '-', color='purple', label='Admirer')
    plot(D, ':', color='yellow', label='Devote')
    decorate(xlabel='Time (days)',
             ylabel='Number of People')
```

```python
frame = song.results
plot_results(frame.N, frame.E, frame.F, frame.A, frame.D)
```
